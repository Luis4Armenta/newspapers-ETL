## Data wrangling

```python
# 1. Añadir newspaper_uid al DataFrame

elsoldemexico['newspaper_uid'] = 'elsoldemexico'

elsoldemexico.head()
```

```python
# 2. Obtener el host
from urllib.parse import urlparse

elsoldemexico['host'] = elsoldemexico['url'].apply(lambda url: urlparse(url).netloc)

elsoldemexico.head()
```

```python
elsoldemexico['host'].value_counts()
```

## Missing data

```python
# 3. Rellenar datos faltantes

missing_titles_mask = elsoldemexico['title'].isna()

missing_titles = (elsoldemexico[missing_titles_mask]['url']
                    .str.extract(r'(?P<missing_titles>[^/]+)$')
                    .applymap(lambda title: title.split('-'))
                    .applymap(lambda title_word_list: ' '.join(title_word_list)))

missing_titles
```

## Additional cleanup

```python
# 4. Añadir uid a las filas
import hashlib

uids = (
  elsoldemexico
    .apply(lambda row: hashlib.md5(bytes(row['url'].encode())), axis=1)
    .apply(lambda hash_object: hash_object.hexdigest())
)

elsoldemexico['uid'] = uids
elsoldemexico.set_index('uid', inplace=True)

elsoldemexico
```

```python
stripped_body = (
  elsoldemexico
    .apply(lambda row: row['body'], axis=1)
    .apply(lambda body: list(body))
    .apply(lambda letters: list(map(lambda letter: letter.replace('\n', ' '), letters)))
    .apply(lambda letters: ''.join(letters))
)

stripped_body.head()
```

```python
stripped_title = (
  elsoldemexico
    .apply(lambda row: row['title'], axis=1)
    .apply(lambda title: list(title))
    .apply(lambda letters: list(map(lambda letter: letter.replace('\n', ' '), letters)))
    .apply(lambda letters: ''.join(letters))
    .apply(lambda title: title.strip())
)

stripped_title.head()
```

## Data enrichment

```python
# 6. tokenizar el titullo y el body
import nltk
from nltk.corpus import stopwords

# nltk.download('punkt')
# nltk.download('stopwords')

stop_words = set(stopwords.words('spanish'))

def tokenize_column(df, column_name):
  return (
    df
      .dropna()
      .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
      .apply(lambda tokens: list(filter(lambda token: token.isalpha(),  tokens)))
      .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
      .apply(lambda word_list: list(filter(lambda word: word not in stop_words, word_list)))
      .apply(lambda valid_word_list: len(valid_word_list))
  )

elsoldemexico['n_tokens_title'] = tokenize_column(elsoldemexico, 'title')
elsoldemexico['n_tokens_body'] = tokenize_column(elsoldemexico, 'body')
elsoldemexico.head()
```

## Duplicate values

```python
# 8. Eliminar duplicados

# elsoldemexico['title'].value_counts().mean()
elsoldemexico.drop_duplicates(subset=['title'], keep='first', inplace=True)

elsoldemexico.head()
```
